```python
import pandas as pd
import numpy as np
import os 
import matplotlib.pyplot as plt 
os.chdir(r'C:\Users\lanmo\OneDrive\Документы\hak')
df = pd.read_excel(r"C:\Users\lanmo\Downloads\train.xlsx")
df.drop('Unnamed: 0',axis = 1,inplace = True)
```

#  Заменяем пустые значения

```python
df = df.fillna(0)
def replace(x):
    if x != 0:
        return 1
    else:
        return 0
df['Дата отмены'] = df['Дата отмены'].apply(replace)
```

```python
df_test = pd.read_excel(r"C:\Users\lanmo\Downloads\test.xlsx")
df_test.drop('Unnamed: 0',axis = 1,inplace = True)
```

```python
df = pd.concat([df,df_test],axis = 0)
```

# Работаем с датами

```python
df['weekday'] = df['Дата бронирования'].apply(lambda x : x.strftime('%a'))
df['delta_day'] = (df['Заезд'] - df['Дата бронирования']).dt.days
df['Quartal'] = df['Дата бронирования'].dt.quarter
df['weekend'] = df['weekday'].apply(lambda x : 1 if x in ['Sat','Sun'] else 0)
```

# Удаляем данные

```python
df1 = df.drop(['Дата бронирования','Заезд','Выезд'],axis = 1)
df1.drop(['weekday','Статус брони','№ брони'],axis = 1,inplace = True)
```

# Работаем с источником

```python
df1.rename(columns = {'Источник':'Source'}, inplace = True )
s = df1.Source.value_counts()
df1['source_count'] = df1['Source'].map(s)
df1.loc[df1['Source'] == "Официальный сайт", "Source"] = "official_website"
df1.loc[df1['Source'] == "booking.com (Booking.com)", "Source"] = "booking.com"
df1.loc[df1['Source'] == "Bronevik.com/Bro.Online", "Source"] = "Bronevik.com(new)"
df1.loc[df1['source_count'] <= 200, 'Source'] = 'small_online_service'
```

# Удаляем пустые значения

```python
df1.drop(['source_count'],axis = 1,inplace =True)
```

# Процент по стоимости

```python
df1['% внесена предоплата к стоимости'] = (df['Внесена предоплата']/df['Стоимость']).apply(lambda x:round(x,2))
df1.drop('Внесена предоплата',axis = 1,inplace = True)
```

```python
df2 = df1
```

```python
df2['Регионы'] = df2['Гостиница'].apply(lambda x: 0 if x in [1,2] else 1)
```

```python
df2['Стоимость на человека'] = (df2['Стоимость']/df2['Гостей']).apply(lambda x:round(x,2))
df2.drop(['Гостей','Гостиница'],axis=1,inplace = True)
```

# Категория номера

```python
def cat(x):
    if '\n' in x:
        return 'Несколько'
    else:
        return x
df2['Категория номера'] = df2['Категория номера'].apply(cat)
```

# Способ оплаты

```python
def sposob(x):
    bank_card_list = [
    "Банк. карта: Банк Россия (банк. карта)",
    "Банк. карта [Кешбэк. МИР]: Эквайринг TravelLine Pro (Банк. карта)",
    "Банк. карта: Эквайринг ComfortBooking (Банк. карта)",
    "Банк. карта [Кешбэк. МИР]: Эквайринг ComfortBooking (Банк. карта)",
    "Банк. карта (SberPay): Эквайринг ComfortBooking (Банк. карта) (SberPay)",
    "Банк. карта (Yandex Pay): Эквайринг ComfortBooking (Банк. карта) (Yandex Pay)",
    "Внешняя система оплаты (Банковская карта)",
    "Гарантия банковской картой"
]
    if x in bank_card_list:
        return 'Банковские карты'
    else:
        return x
df2['Способ оплаты'] = df2['Способ оплаты'].apply(sposob)
```
